## Part 1: GridWorld Environment

We'll use a **5x5 GridWorld** with features that make learning interesting:
- **Walls** (black) at (1,1), (2,3), (3,1) — block movement (agent stays in place)
- **Traps** (orange) at (1,3), (3,2) — give **-5 penalty** each visit
- **Goal** (gold) at (4,4) — gives **+10 reward** and ends the episode
- All other steps cost **-1** (encourages finding short paths)

```python
class GridWorld:
    """5x5 GridWorld with walls and traps"""
    def __init__(self, size=5):
        self.size = size
        self.start_state = (0, 0)
        self.goal_state = (size - 1, size - 1)
        self.walls = [(1, 1), (2, 3), (3, 1)]
        self.traps = [(1, 3), (3, 2)]
        self.reset()

    def reset(self):
        self.state = self.start_state
        return self.state

    def step(self, action):
        """Actions: 0=up, 1=right, 2=down, 3=left"""
        x, y = self.state

        if action == 0:    # up
            nx, ny = max(0, x - 1), y
        elif action == 1:  # right
            nx, ny = x, min(self.size - 1, y + 1)
        elif action == 2:  # down
            nx, ny = min(self.size - 1, x + 1), y
        elif action == 3:  # left
            nx, ny = x, max(0, y - 1)

        # Walls block movement — agent stays in place
        if (nx, ny) in self.walls:
            nx, ny = x, y

        self.state = (nx, ny)

        # Reward structure
        if self.state == self.goal_state:
            return self.state, 10.0, True
        elif self.state in self.traps:
            return self.state, -5.0, False
        else:
            return self.state, -1.0, False

    def render(self, Q=None, title="GridWorld"):
        """Visualize the grid and optionally the learned policy"""
        fig, ax = plt.subplots(figsize=(8, 8))
        action_arrows = {0: '^', 1: '>', 2: 'v', 3: '<'}

        for i in range(self.size):
            for j in range(self.size):
                if (i, j) in self.walls:
                    ax.add_patch(plt.Rectangle((j, i), 1, 1, facecolor='black'))
                    ax.text(j + 0.5, i + 0.5, 'WALL', ha='center', va='center',
                            fontsize=9, color='white', weight='bold')
                elif (i, j) == self.goal_state:
                    ax.add_patch(plt.Rectangle((j, i), 1, 1, facecolor='gold', edgecolor='black'))
                    ax.text(j + 0.5, i + 0.5, 'GOAL\n+10', ha='center', va='center',
                            fontsize=10, weight='bold')
                elif (i, j) in self.traps:
                    ax.add_patch(plt.Rectangle((j, i), 1, 1, facecolor='darkorange', edgecolor='black'))
                    ax.text(j + 0.5, i + 0.3, 'TRAP\n-5', ha='center', va='center',
                            fontsize=9, weight='bold', color='white')
                    if Q is not None and (i, j) in Q:
                        q_values = Q[(i, j)]
                        if isinstance(q_values, torch.Tensor):
                            q_values = q_values.cpu().numpy()
                        best_action = np.argmax(q_values)
                        ax.text(j + 0.5, i + 0.7, action_arrows[best_action],
                                ha='center', va='center', fontsize=16, color='white')
                else:
                    ax.add_patch(plt.Rectangle((j, i), 1, 1, facecolor='lightgray', edgecolor='black'))
                    if Q is not None and (i, j) in Q:
                        q_values = Q[(i, j)]
                        if isinstance(q_values, torch.Tensor):
                            q_values = q_values.cpu().numpy()
                        best_action = np.argmax(q_values)
                        best_value = q_values[best_action]
                        ax.text(j + 0.5, i + 0.4, action_arrows[best_action],
                                ha='center', va='center', fontsize=20, color='blue')
                        ax.text(j + 0.5, i + 0.75, f'{best_value:.1f}',
                                ha='center', va='center', fontsize=9, style='italic')

        # Mark start cell
        ax.text(0.5, 0.05, 'START', ha='center', va='top', fontsize=7, color='gray')

        ax.set_xlim(0, self.size)
        ax.set_ylim(0, self.size)
        ax.set_aspect('equal')
        ax.invert_yaxis()
        ax.set_xticks([])
        ax.set_yticks([])
        plt.title(title)
        plt.tight_layout()
        return fig

# Test the environment
env = GridWorld()
env.render()
plt.show()
```

## Part 2: Monte Carlo Control (❗ solution updates here)

### Theory

Monte Carlo (MC) control learns Q-values from **complete episodes**:
- **No bootstrapping**: Updates based on actual returns
- **High variance, zero bias**: Uses full trajectory
- **Epsilon-greedy exploration**: Balance exploration vs exploitation

**Key Update Rule:**
$$Q(s, a) \leftarrow Q(s, a) + \alpha [G_t - Q(s, a)]$$

where $G_t = R_{t+1} + \gamma R_{t+2} + \gamma^2 R_{t+3} + \ldots$ is the return.

**Policy**: $\pi(s) = \begin{cases} \arg\max_a Q(s,a) & \text{with prob } 1-\epsilon \\ \text{random} & \text{with prob } \epsilon \end{cases}$

### Exercise 1: SOLUTION - Monte Carlo

```python
class MonteCarloControl:
    """Monte Carlo Control with epsilon-greedy exploration"""
    def __init__(self, n_actions=4, gamma=0.99, alpha=0.1, epsilon=0.1, use_gpu=True):
        self.n_actions = n_actions
        self.gamma = gamma
        self.alpha = alpha
        self.epsilon = epsilon
        self.device = device if use_gpu else torch.device('cpu')
        self.Q = defaultdict(lambda: torch.zeros(n_actions, device=self.device))
        
    def epsilon_greedy_policy(self, state):
        """Epsilon-greedy action selection"""
        if np.random.random() < self.epsilon:
            return np.random.randint(0, self.n_actions)
        else:
            return self.Q[state].argmax().item()
    
    def greedy_policy(self, state):
        """Greedy action selection (for evaluation)"""
        return self.Q[state].argmax().item()
    
    def generate_episode(self, env, max_steps=200):
        """Generate episode using epsilon-greedy policy"""
        episode = []
        state = env.reset()
        for _ in range(max_steps):
            action = self.epsilon_greedy_policy(state)
            next_state, reward, done = env.step(action)
            episode.append((state, action, reward))
            state = next_state
            if done:
                break
        return episode
    
    def update(self, episode):
        """Update Q-values using first-visit MC (Buggy version)"""
        G = 0.0
        visited = set()
        
        for state, action, reward in reversed(episode):
            G = reward + self.gamma * G
            
            sa_pair = (state, action)
            if sa_pair not in visited:
                visited.add(sa_pair)
                old_q = self.Q[state][action].item()
                self.Q[state][action] += self.alpha * (G - old_q)
    
    def train(self, env, num_episodes=1000):
        """Train using Monte Carlo control"""
        episode_rewards = []
        for ep in tqdm(range(num_episodes), desc="MC Control Training"):
            episode = self.generate_episode(env)
            self.update(episode)
            total_reward = sum([r for _, _, r in episode])
            episode_rewards.append(total_reward)
            self.epsilon = max(0.01, self.epsilon * 0.995)
        return episode_rewards
```

The code given in the previous solution is not implementing first-visit Monte Carlo. Because the episode is processed in reverse order, the first time a state-action pair is encountered in the loop corresponds to its last occurrence in the original episode. So this results in a last-visit update rather than a first-visit update.

The reason the previous version was written this way is that iterating backward allows us to compute the return `G` efficiently in linear time. If one instead loops forward and recomputes the full return from each time step naively, this leads to quadratic complexity in the episode length. So the backward pass was meant to avoid an O(n^2) computation of the returns, but the visited logic was unfortunately not consistent with true first-visit MC.

A corrected version that keeps the return computation linear is the following:

```python
class MonteCarloControlFixed(MonteCarloControl):
    """Monte Carlo Control with correct first-visit MC update"""
    
    def update(self, episode):
        """Update Q-values using correct first-visit MC.
        
        Iterates in reverse (O(n)) to compute returns efficiently.
        Earlier visits overwrite later ones in the dict, so after the loop
        each (state, action) holds the return from its first visit.
        """
        G = 0.0
        first_visit_returns = {}
        
        for state, action, reward in reversed(episode):
            G = reward + self.gamma * G
            first_visit_returns[(state, action)] = G  # last written in reverse = first chronologically
        
        for (state, action), G in first_visit_returns.items():
            old_q = self.Q[state][action].item()
            self.Q[state][action] += self.alpha * (G - old_q)
```

```python
# Compare original (last-visit) vs fixed (first-visit) MC
print('Training original MC (last-visit)...')
env_orig = GridWorld()
mc_orig = MonteCarloControl(gamma=0.95, alpha=0.1, epsilon=0.3)
rewards_mc_orig = mc_orig.train(env_orig, num_episodes=5000)

print('Training fixed MC (first-visit)...')
env = GridWorld()
mc = MonteCarloControlFixed(gamma=0.95, alpha=0.1, epsilon=0.3)
rewards_mc = mc.train(env, num_episodes=5000)

# Learning curves side by side
fig, axes = plt.subplots(1, 2, figsize=(16, 4))
for ax, rewards, label in [
    (axes[0], rewards_mc_orig, 'Original MC (last-visit, buggy)'),
    (axes[1], rewards_mc, 'Fixed MC (first-visit, correct)'),
]:
    ax.plot(rewards, alpha=0.3)
    ax.plot(np.convolve(rewards, np.ones(50)/50, mode='valid'), linewidth=2)
    ax.set_xlabel('Episode')
    ax.set_ylabel('Total Reward')
    ax.set_title(label)
    ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# Policy comparison
env_orig.render(Q=mc_orig.Q, title='Learned Policy (Original MC)')
plt.show()
env.render(Q=mc.Q, title='Learned Policy (Fixed MC)')
plt.show()
```

## Theory Check 1: Monte Carlo vs TD Learning - SOLUTIONS

### Q1: Monte Carlo - Zero Bias, High Variance

**(a) Zero bias because:** MC uses the actual return G (sum of real rewards) to update Q-values. Since G is an unbiased sample of the true expected return, the MC estimate converges to the true value without systematic error.

**(b) High variance because:** G depends on the entire sequence of future rewards and actions. Each episode can follow a very different trajectory, so G fluctuates widely from episode to episode. More random variables in the sum = more variance.

### Q2: Calculate TD Update

- **(a) TD target = r + gamma x Q(s',a') = 5 + 0.9 x 8 = 12.2**
- **(b) TD error = TD target - Q(s,a) = 12.2 - 50 = -37.8**
- **(c) New Q(s,a) = 50 + 0.1 x (-37.8) = 46.22**

### Q3: TD Variance

TD depends on only **1 random reward** and **1 Q-value estimate**. MC depends on **all future rewards** (many random variables summed together). Fewer random variables -> lower variance. The tradeoff is that TD introduces some bias because it bootstraps from an estimated Q-value rather than using the true return.

## Part 3: SARSA (On-Policy TD Control)

### Theory

SARSA (State-Action-Reward-State-Action) is an **on-policy** TD control algorithm:
- Learns Q-values for the policy being followed
- Uses bootstrapping (updates from estimates)
- More conservative than Q-learning

**SARSA Update:**
$$Q(s_t, a_t) \leftarrow Q(s_t, a_t) + \alpha [r_{t+1} + \gamma Q(s_{t+1}, a_{t+1}) - Q(s_t, a_t)]$$

**Key**: Uses actual next action $a_{t+1}$ sampled from policy (e.g., $\epsilon$-greedy).

### Comparison with MC:
- **MC**: Waits for episode end, uses full return G
- **SARSA**: Updates every step, uses one-step lookahead

### Exercise 2: SOLUTION - SARSA

```python
class SARSA:
    """SARSA: On-policy TD Control"""
    def __init__(self, n_actions=4, gamma=0.99, alpha=0.1, epsilon=0.1, use_gpu=True):
        self.n_actions = n_actions
        self.gamma = gamma
        self.alpha = alpha
        self.epsilon = epsilon
        self.device = device if use_gpu else torch.device('cpu')
        self.Q = defaultdict(lambda: torch.zeros(n_actions, device=self.device))
        
    def epsilon_greedy(self, state):
        """Epsilon-greedy policy"""
        if np.random.random() < self.epsilon:
            return np.random.randint(0, self.n_actions)
        return self.Q[state].argmax().item()
    
    def greedy_policy(self, state):
        return self.Q[state].argmax().item()
    
    def update(self, state, action, reward, next_state, next_action, done):
        """SARSA update rule"""
        current_q = self.Q[state][action]
        
        if done:
            td_target = reward
        else:
            # Use Q-value of NEXT action (from policy) - ON-POLICY
            td_target = reward + self.gamma * self.Q[next_state][next_action]
        
        td_error = td_target - current_q
        self.Q[state][action] += self.alpha * td_error
    
    def train(self, env, num_episodes=1000, max_steps=100):
        """Train SARSA agent"""
        episode_rewards = []
        for ep in tqdm(range(num_episodes), desc="SARSA Training"):
            state = env.reset()
            action = self.epsilon_greedy(state)
            total_reward = 0
            for step in range(max_steps):
                next_state, reward, done = env.step(action)
                next_action = self.epsilon_greedy(next_state)
                self.update(state, action, reward, next_state, next_action, done)
                total_reward += reward
                state, action = next_state, next_action
                if done:
                    break
            episode_rewards.append(total_reward)
            self.epsilon = max(0.01, self.epsilon * 0.995)
        return episode_rewards
```

```python
# Train SARSA
print('Training SARSA...')
env = GridWorld()
sarsa = SARSA(gamma=0.95, alpha=0.1, epsilon=0.3)
rewards_sarsa = sarsa.train(env, num_episodes=3000)

plt.figure(figsize=(12, 4))
plt.plot(rewards_sarsa, alpha=0.3)
plt.plot(np.convolve(rewards_sarsa, np.ones(50)/50, mode='valid'), linewidth=2)
plt.xlabel('Episode')
plt.ylabel('Total Reward')
plt.title('SARSA Learning Progress')
plt.grid(True, alpha=0.3)
plt.show()

env.render(Q=sarsa.Q, title='Learned Policy (SARSA)')
plt.show()
```

## Part 4: Q-Learning (Off-Policy TD Control)

### Theory

Q-Learning is an **off-policy** TD control algorithm:
- Learns optimal Q-values regardless of behavior policy
- Uses max over next actions (greedy)
- More aggressive than SARSA

**Q-Learning Update:**
$$Q(s_t, a_t) \leftarrow Q(s_t, a_t) + \alpha [r_{t+1} + \gamma \max_a Q(s_{t+1}, a) - Q(s_t, a_t)]$$

**Key difference from SARSA**: Uses $\max_a Q(s_{t+1}, a)$ instead of $Q(s_{t+1}, a_{t+1})$.

This means Q-learning learns the **optimal policy** even while exploring!

### Exercise 3: SOLUTION - Q-Learning

```python
class QLearning:
    """Q-Learning: Off-policy TD Control"""
    def __init__(self, n_actions=4, gamma=0.99, alpha=0.1, epsilon=0.1, use_gpu=True):
        self.n_actions = n_actions
        self.gamma = gamma
        self.alpha = alpha
        self.epsilon = epsilon
        self.device = device if use_gpu else torch.device('cpu')
        self.Q = defaultdict(lambda: torch.zeros(n_actions, device=self.device))
        
    def epsilon_greedy(self, state):
        """Behavior policy (epsilon-greedy)"""
        if np.random.random() < self.epsilon:
            return np.random.randint(0, self.n_actions)
        return self.Q[state].argmax().item()
    
    def greedy_policy(self, state):
        """Target policy (greedy)"""
        return self.Q[state].argmax().item()
    
    def update(self, state, action, reward, next_state, done):
        """Q-Learning update rule"""
        current_q = self.Q[state][action]
        
        if done:
            td_target = reward
        else:
            # Use MAX Q-value (greedy) - OFF-POLICY
            td_target = reward + self.gamma * self.Q[next_state].max()
        
        td_error = td_target - current_q
        self.Q[state][action] += self.alpha * td_error
    
    def train(self, env, num_episodes=1000, max_steps=100):
        """Train Q-Learning agent"""
        episode_rewards = []
        for ep in tqdm(range(num_episodes), desc="Q-Learning Training"):
            state = env.reset()
            total_reward = 0
            for step in range(max_steps):
                action = self.epsilon_greedy(state)
                next_state, reward, done = env.step(action)
                self.update(state, action, reward, next_state, done)
                total_reward += reward
                state = next_state
                if done:
                    break
            episode_rewards.append(total_reward)
            self.epsilon = max(0.01, self.epsilon * 0.995)
        return episode_rewards
```

```python
# Train Q-Learning
print('Training Q-Learning...')
env = GridWorld()
qlearning = QLearning(gamma=0.95, alpha=0.1, epsilon=0.3)
rewards_qlearning = qlearning.train(env, num_episodes=3000)

plt.figure(figsize=(12, 4))
plt.plot(rewards_qlearning, alpha=0.3)
plt.plot(np.convolve(rewards_qlearning, np.ones(50)/50, mode='valid'), linewidth=2)
plt.xlabel('Episode')
plt.ylabel('Total Reward')
plt.title('Q-Learning Progress')
plt.grid(True, alpha=0.3)
plt.show()

env.render(Q=qlearning.Q, title='Learned Policy (Q-Learning)')
plt.show()
```

## (❗ New) Part 5b: Does grid size affect MC vs TD convergence?

The 5x5 grid has short episodes, which reduces MC's variance disadvantage.
A 10x10 grid with more obstacles forces longer paths — let's see if TD pulls ahead.

```python
class GridWorld10x10(GridWorld):
    """10x10 GridWorld — two partial barriers force longer paths than 5x5."""
    def __init__(self):
        self.size = 10
        self.start_state = (0, 0)
        self.goal_state = (9, 9)
        self.walls = [
            (2, 2), (2, 3), (2, 4), (2, 5), (2, 6),
            (6, 3), (6, 4), (6, 5), (6, 6), (6, 7),
            (4, 1), (4, 8),
        ]
        self.traps = [(1, 6), (3, 2), (4, 5), (5, 8), (7, 3), (8, 6)]
        self.reset()


def evaluate_policy(agent, n_eval=50, max_steps=500):
    """Evaluate greedy policy over n_eval episodes — no exploration."""
    total = 0
    for _ in range(n_eval):
        env = GridWorld10x10()
        state = env.reset()
        ep_reward = 0
        for _ in range(max_steps):
            action = agent.greedy_policy(state)
            state, reward, done = env.step(action)
            ep_reward += reward
            if done:
                break
        total += ep_reward
    return total / n_eval


def train_mc_with_eval(agent, eval_every=500, total_steps=1500000):
    """Train MC, evaluating every eval_every environment steps."""
    env = GridWorld10x10()
    eval_rewards, eval_steps = [], []
    steps = 0
    next_eval = eval_every
    pbar = tqdm(total=total_steps, desc='MC Training (steps)')
    while steps < total_steps:
        episode = agent.generate_episode(env, max_steps=1000)
        agent.update(episode)
        agent.epsilon = max(0.01, agent.epsilon * 0.9995)
        steps += len(episode)
        pbar.update(len(episode))
        if steps >= next_eval:
            eval_rewards.append(evaluate_policy(agent))
            eval_steps.append(steps)
            next_eval += eval_every
    pbar.close()
    return eval_steps, eval_rewards


def train_sarsa_with_eval(agent, eval_every=500, total_steps=1500000, max_steps=100):
    """Train SARSA, evaluating every eval_every environment steps."""
    env = GridWorld10x10()
    eval_rewards, eval_steps = [], []
    steps = 0
    next_eval = eval_every
    pbar = tqdm(total=total_steps, desc='SARSA Training (steps)')
    while steps < total_steps:
        state = env.reset()
        action = agent.epsilon_greedy(state)
        for _ in range(max_steps):
            next_state, reward, done = env.step(action)
            next_action = agent.epsilon_greedy(next_state)
            agent.update(state, action, reward, next_state, next_action, done)
            state, action = next_state, next_action
            steps += 1
            pbar.update(1)
            if done:
                break
        agent.epsilon = max(0.01, agent.epsilon * 0.995)
        if steps >= next_eval:
            eval_rewards.append(evaluate_policy(agent))
            eval_steps.append(steps)
            next_eval += eval_every
    pbar.close()
    return eval_steps, eval_rewards


def train_qlearning_with_eval(agent, eval_every=500, total_steps=1500000, max_steps=100):
    """Train Q-Learning, evaluating every eval_every environment steps."""
    env = GridWorld10x10()
    eval_rewards, eval_steps = [], []
    steps = 0
    next_eval = eval_every
    pbar = tqdm(total=total_steps, desc='Q-Learning Training (steps)')
    while steps < total_steps:
        state = env.reset()
        for _ in range(max_steps):
            action = agent.epsilon_greedy(state)
            next_state, reward, done = env.step(action)
            agent.update(state, action, reward, next_state, done)
            state = next_state
            steps += 1
            pbar.update(1)
            if done:
                break
        agent.epsilon = max(0.01, agent.epsilon * 0.995)
        if steps >= next_eval:
            eval_rewards.append(evaluate_policy(agent))
            eval_steps.append(steps)
            next_eval += eval_every
    pbar.close()
    return eval_steps, eval_rewards


# Visualise grid
env10 = GridWorld10x10()
env10.render(title='10x10 GridWorld')
plt.show()

# Train all three — same total environment steps for fair comparison
TOTAL_STEPS = 100_000
TOTAL_STEPS_QUICK = 500_00  # For quick testing, reduce this to 100k steps
mc10    = MonteCarloControlFixed(gamma=0.99, alpha=0.05, epsilon=0.3)
sarsa10 = SARSA(gamma=0.99, alpha=0.1, epsilon=0.3)
ql10    = QLearning(gamma=0.99, alpha=0.1, epsilon=0.3)

steps_mc,    r_mc    = train_mc_with_eval(mc10,    total_steps=TOTAL_STEPS)
steps_sarsa, r_sarsa = train_sarsa_with_eval(sarsa10, total_steps=TOTAL_STEPS_QUICK)
steps_ql,    r_ql    = train_qlearning_with_eval(ql10, total_steps=TOTAL_STEPS_QUICK)

# Plot greedy policy performance vs total environment steps
plt.figure(figsize=(12, 5))
plt.plot(steps_mc,    r_mc,    label='MC (first-visit)', linewidth=2)
plt.plot(steps_sarsa, r_sarsa, label='SARSA',            linewidth=2)
plt.plot(steps_ql,    r_ql,    label='Q-Learning',       linewidth=2)
plt.xlabel('Total Environment Steps')
plt.ylabel('Avg Greedy Reward (50 eval episodes)')
plt.title('10x10 GridWorld: Sample Efficiency (steps-matched)')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print('\n' + '='*50)
print('FINAL GREEDY POLICY PERFORMANCE')
print('='*50)
print(f"{'MC (first-visit)':<25} {r_mc[-1]:>10.2f}")
print(f"{'SARSA':<25} {r_sarsa[-1]:>10.2f}")
print(f"{'Q-Learning':<25} {r_ql[-1]:>10.2f}")
print('='*50)
```

### Section 5b Summary: What Does Grid Size Reveal About MC vs TD?

#### On the 5x5 grid

Fixed first-visit MC is the **correct** algorithm and it shows: compared to the buggy last-visit variant, it converges more reliably and reaches a higher average reward. That said, all three methods — MC, SARSA, and Q-Learning — eventually find a good policy on the 5x5 grid. The episodes are short, returns are low-variance, and the state space is small enough that MC's weaknesses don't really hurt it.

#### On the 10x10 grid (Part 5b / 6b experiment)

The larger grid exposes a practical limitation of Monte Carlo: **it requires noticeably more tuning to work well**. To get MC performing competitively, we had to make several adjustments that were not needed for SARSA or Q-Learning:

| Hyperparameter | MC (10×10) | SARSA / Q-Learning (10×10) |
|---|---|---|
| Learning rate α | 0.05 (halved) | 0.1 (unchanged from 5×5) |
| Epsilon decay | 0.9995 (slower) | 0.995 (unchanged) |
| Max steps per episode | 1000 | 100 |
| Total environment steps | 100 000 (2×) | 50 000 |

SARSA and Q-Learning worked straight away with the same hyperparameters used on the 5×5 grid.

#### Interpreting the 10x10 plot

The learning curve (greedy policy reward vs total environment steps) tells a clear story:

- **SARSA and Q-Learning improve quickly**, reaching a stable policy well within their 50 000-step budget.
- **MC lags behind** in the early phase — it needs longer episodes and more data before returns are informative enough to drive useful updates.
- Ultimately, all three agents converge to a final greedy reward of **−7**, meaning they find policies of comparable quality. However, the TD methods got there with *half* the experience and with no special tuning.

#### Why does this happen?

In a larger grid, episodes are longer and the return at each step depends on many more future rewards. This **inflates MC's variance** — each estimate of (s,a) is noisier, so learning slows down. TD methods sidestep this by bootstrapping: they update from a single reward plus an estimated future value, keeping variance low regardless of episode length. The cost is a small amount of bias (because they use estimates, not true returns) — but on typical problems that tradeoff is well worth it.

> **Key takeaway:** MC and TD all converge to the same policy eventually, but TD methods are far more *sample-efficient* and *robust to grid scale*. As environments grow, the variance problem of MC compounds, while TD methods scale gracefully with minimal retuning.

## Part 6: SARSA vs Q-Learning on Cliff Walking

The classic **Cliff Walking** problem (Sutton & Barto, Example 6.6) demonstrates the key difference between on-policy (SARSA) and off-policy (Q-Learning).

**Layout**: 4x10 grid. Start = bottom-left, Goal = bottom-right, Cliff = entire bottom row between them.
- Stepping on the cliff gives **-100** reward and resets to start
- The **optimal path** walks right along the cliff edge (shortest)
- The **safe path** goes up and across the top (longer but no cliff risk)

**Key insight**: Q-Learning learns the optimal (risky) path. SARSA learns the safe path because it accounts for the fact that epsilon-greedy exploration can randomly step into the cliff.

```python
class CliffWalking:
    """Classic Cliff Walking (Sutton & Barto Example 6.6)
    4x10 grid. Start=bottom-left, Goal=bottom-right.
    Cliff along the bottom row between start and goal."""
    def __init__(self):
        self.rows = 4
        self.cols = 10
        self.start = (3, 0)
        self.goal = (3, 9)
        # Cliff: bottom row columns 1-8
        self.cliff = [(3, c) for c in range(1, 9)]
        self.reset()

    def reset(self):
        self.state = self.start
        return self.state

    def step(self, action):
        """Actions: 0=up, 1=right, 2=down, 3=left"""
        x, y = self.state

        if action == 0:
            x = max(0, x - 1)
        elif action == 1:
            y = min(self.cols - 1, y + 1)
        elif action == 2:
            x = min(self.rows - 1, x + 1)
        elif action == 3:
            y = max(0, y - 1)

        self.state = (x, y)

        if self.state in self.cliff:
            self.state = self.start
            return self.state, -100, False
        elif self.state == self.goal:
            return self.state, 0, True
        else:
            return self.state, -1, False

    def render(self, path=None, title="Cliff Walking"):
        """Visualize the cliff environment"""
        fig, ax = plt.subplots(figsize=(12, 5))

        for i in range(self.rows):
            for j in range(self.cols):
                if (i, j) == self.start:
                    ax.add_patch(plt.Rectangle((j, i), 1, 1, facecolor='green', edgecolor='black'))
                    ax.text(j + 0.5, i + 0.5, 'S', ha='center', va='center', fontsize=12, weight='bold')
                elif (i, j) == self.goal:
                    ax.add_patch(plt.Rectangle((j, i), 1, 1, facecolor='gold', edgecolor='black'))
                    ax.text(j + 0.5, i + 0.5, 'G', ha='center', va='center', fontsize=12, weight='bold')
                elif (i, j) in self.cliff:
                    ax.add_patch(plt.Rectangle((j, i), 1, 1, facecolor='red', edgecolor='black'))
                    ax.text(j + 0.5, i + 0.5, 'CLIFF', ha='center', va='center', fontsize=7, color='white')
                else:
                    ax.add_patch(plt.Rectangle((j, i), 1, 1, facecolor='lightgray', edgecolor='black'))

        if path:
            path_x = [p[1] + 0.5 for p in path]
            path_y = [p[0] + 0.5 for p in path]
            ax.plot(path_x, path_y, 'b-', linewidth=3, alpha=0.6, marker='o', markersize=5)

        ax.set_xlim(0, self.cols)
        ax.set_ylim(0, self.rows)
        ax.set_aspect('equal')
        ax.invert_yaxis()
        ax.set_xticks([])
        ax.set_yticks([])
        ax.set_title(title)
        return fig

# Visualize the environment
cliff_env = CliffWalking()
cliff_env.render()
plt.show()

# Train with fixed epsilon to get comparable reward curves
# Fixed epsilon is critical: the SARSA vs Q-Learning difference only shows
# when the agent keeps exploring (epsilon > 0) throughout training.
sarsa_cliff = SARSA(gamma=0.99, alpha=0.5, epsilon=0.1)
ql_cliff = QLearning(gamma=0.99, alpha=0.5, epsilon=0.1)

rewards_sarsa_cliff = []
rewards_ql_cliff = []
for ep in tqdm(range(500), desc="Cliff Walking Training"):
    # SARSA episode
    env = CliffWalking()
    state = env.reset()
    action = sarsa_cliff.epsilon_greedy(state)
    total_r = 0
    for _ in range(200):
        ns, r, done = env.step(action)
        na = sarsa_cliff.epsilon_greedy(ns)
        sarsa_cliff.update(state, action, r, ns, na, done)
        total_r += r
        state, action = ns, na
        if done: break
    rewards_sarsa_cliff.append(total_r)
    sarsa_cliff.epsilon = 0.1  # keep fixed

    # Q-Learning episode
    env = CliffWalking()
    state = env.reset()
    total_r = 0
    for _ in range(200):
        action = ql_cliff.epsilon_greedy(state)
        ns, r, done = env.step(action)
        ql_cliff.update(state, action, r, ns, done)
        total_r += r
        state = ns
        if done: break
    rewards_ql_cliff.append(total_r)
    ql_cliff.epsilon = 0.1  # keep fixed

# Compare performance
plt.figure(figsize=(12, 5))
window = 20
plt.plot(np.convolve(rewards_sarsa_cliff, np.ones(window)/window, mode='valid'),
         label='SARSA (safer path)', linewidth=2, color='blue')
plt.plot(np.convolve(rewards_ql_cliff, np.ones(window)/window, mode='valid'),
         label='Q-Learning (optimal but risky)', linewidth=2, color='red')
plt.xlabel('Episode')
plt.ylabel('Reward (Moving Average)')
plt.title('SARSA vs Q-Learning on Cliff Walking')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# Show learned paths (greedy, no exploration)
def get_path(agent, env, max_steps=50):
    """Extract greedy path from learned policy"""
    path = []
    state = env.reset()
    path.append(state)
    for _ in range(max_steps):
        action = agent.greedy_policy(state)
        state, _, done = env.step(action)
        path.append(state)
        if done:
            break
    return path

sarsa_path = get_path(sarsa_cliff, CliffWalking())
CliffWalking().render(path=sarsa_path, title='SARSA: Safe Path (avoids cliff edge)')
plt.show()

ql_path = get_path(ql_cliff, CliffWalking())
CliffWalking().render(path=ql_path, title='Q-Learning: Optimal Path (along cliff edge)')
plt.show()

print("\n" + "="*70)
print("KEY DIFFERENCE: On-Policy (SARSA) vs Off-Policy (Q-Learning)")
print("="*70)
print("SARSA (On-Policy):")
print("  - Learns value of policy being followed (epsilon-greedy)")
print("  - Takes SAFER path away from cliff edge")
print("  - Higher average reward during training (fewer cliff falls)")
print("")
print("Q-Learning (Off-Policy):")
print("  - Learns optimal policy regardless of behavior")
print("  - Takes OPTIMAL path right along cliff edge")
print("  - Lower average reward during training (cliff falls from exploration)")
print("="*70)
```

**Q4:** Use **SARSA** (safety-critical). SARSA accounts for exploration risk in its Q-values. Since the robot will sometimes explore randomly near expensive equipment, SARSA learns to stay away. Q-Learning would learn the optimal path that may pass dangerously close.

**Q6 (Math):** Let the return samples for a fixed $(s,a)$ under policy $\pi$ be $G_1,\dots,G_n$ with
$\mathbb{E}[G_i]=q_\pi(s,a)$ and $\mathrm{Var}(G_i)=\sigma^2<\infty$.
Define
$$\hat q_n(s,a)=\frac{1}{n}\sum_{i=1}^n G_i.$$
Show:
1. $\mathbb{E}[\hat q_n(s,a)] = q_\pi(s,a)$
2. $\mathrm{Var}(\hat q_n(s,a)) = \sigma^2/n$
3. Use Chebyshev's inequality to conclude $\hat q_n(s,a) \xrightarrow{P} q_\pi(s,a)$ as $n\to\infty$.

**Answer:**
1. By linearity of expectation,
$$\mathbb{E}[\hat q_n]=\frac{1}{n}\sum_{i=1}^n \mathbb{E}[G_i]
=\frac{1}{n}\sum_{i=1}^n q_\pi(s,a)=q_\pi(s,a).$$
So the sample-mean Monte Carlo estimator is unbiased.

2. Assuming independent episode returns,
$$\mathrm{Var}(\hat q_n)=\mathrm{Var}\left(\frac{1}{n}\sum_{i=1}^n G_i\right)
=\frac{1}{n^2}\sum_{i=1}^n \mathrm{Var}(G_i)=\frac{n\sigma^2}{n^2}=\frac{\sigma^2}{n}.$$

3. For any $\varepsilon>0$, Chebyshev gives
$$\Pr\left(|\hat q_n-q_\pi|\ge\varepsilon\right)
\le \frac{\mathrm{Var}(\hat q_n)}{\varepsilon^2}
=\frac{\sigma^2}{n\varepsilon^2}\to 0.$$
Hence $\hat q_n \xrightarrow{P} q_\pi$.

**Q7 (Math - Q-Learning Convergence):** For tabular Q-learning,
$$Q_{t+1}(s_t,a_t)=Q_t(s_t,a_t)+\alpha_t\big[r_t+\gamma\max_{a'}Q_t(s_{t+1},a')-Q_t(s_t,a_t)\big],\quad 0\le\gamma<1.$$

(a) Prove in detail that the Bellman optimality operator
$$ (T^*Q)(s,a)=\mathbb{E}[r+\gamma\max_{a'}Q(s',a')\mid s,a] $$
is a $\gamma$-contraction in sup norm.


**Answer:**
(a) Detailed contraction proof:
Take any two action-value functions $Q,\bar Q$. For any fixed $(s,a)$,
\begin{align}
|(T^*Q)(s,a)-(T^*\bar Q)(s,a)|
&=\left|\mathbb E\left[\gamma\Big(\max_{a'}Q(s',a')-\max_{a'}\bar Q(s',a')\Big)\mid s,a\right]\right|\\
&\le \gamma\,\mathbb E\left[\left|\max_{a'}Q(s',a')-\max_{a'}\bar Q(s',a')\right|\mid s,a\right].
\end{align}
Now use
$$\left|\max_i x_i-\max_i y_i\right|\le \max_i|x_i-y_i|.$$
So
\begin{align}
|(T^*Q)(s,a)-(T^*\bar Q)(s,a)|
&\le \gamma\,\mathbb E\left[\max_{a'}|Q(s',a')-\bar Q(s',a')|\mid s,a\right]\\
&\le \gamma\,\|Q-\bar Q\|_\infty.
\end{align}
Taking supremum over $(s,a)$ gives
$$\|T^*Q-T^*\bar Q\|_\infty\le \gamma\|Q-\bar Q\|_\infty.$$
So $T^*$ is a $\gamma$-contraction.

Existence of a fixed point (without named theorem): pick any $Q_0$ and define
$$Q_{k+1}=T^*Q_k.$$ 
Then
$$\|Q_{k+1}-Q_k\|_\infty \le \gamma^k\|Q_1-Q_0\|_\infty.$$ 
Hence for $m>n$,
$$\|Q_m-Q_n\|_\infty
\le \sum_{j=n}^{m-1}\|Q_{j+1}-Q_j\|_\infty
\le \frac{\gamma^n}{1-\gamma}\|Q_1-Q_0\|_\infty\to 0,$$
so $\{Q_k\}$ is Cauchy and converges (finite-dimensional space). Let limit be $\bar Q$. Since $T^*$ is Lipschitz, it is continuous, so
$$T^*\bar Q=T^*\left(\lim_k Q_k\right)=\lim_k T^*Q_k=\lim_k Q_{k+1}=\bar Q.$$
Thus a fixed point exists.

Uniqueness: if $Q$ and $Q'$ are fixed points,
$$\|Q-Q'\|_\infty=\|T^*Q-T^*Q'\|_\infty\le \gamma\|Q-Q'\|_\infty.$$ 
With $0\le\gamma<1$, this implies $\|Q-Q'\|_\infty=0$, so $Q=Q'$. Therefore the fixed point is unique, denoted $Q^*$.

## Part 7: Additional Material - Bias-Variance Analysis

Let's empirically analyze the bias-variance tradeoff between MC and TD methods.

```python
def compare_mc_td_variance(n_runs=25, n_episodes=2000):
    """Compare variance of MC vs TD Q-value estimates"""
    
    target_state_action = ((1, 2), 1)
    state, action = target_state_action

    print('Computing ground-truth Q-value (long Q-Learning run)...')
    true_agent = QLearning(gamma=0.95, alpha=0.05, epsilon=0.1)
    true_agent.train(GridWorld(), num_episodes=50000)
    true_q = true_agent.Q[state][action].item()
    print(f'Ground-truth Q{target_state_action} ≈ {true_q:.4f}\n')

    mc_orig_estimates = []
    mc_fixed_estimates = []
    td_estimates = []
    
    print(f'Running {n_runs} independent trials ({n_episodes} episodes each)...')
    for run in tqdm(range(n_runs)):
        mc_o = MonteCarloControl(gamma=0.95, alpha=0.1, epsilon=0.1)
        mc_o.train(GridWorld(), num_episodes=n_episodes)
        mc_orig_estimates.append(mc_o.Q[state][action].item())

        mc_f = MonteCarloControlFixed(gamma=0.95, alpha=0.1, epsilon=0.1)
        mc_f.train(GridWorld(), num_episodes=n_episodes)
        mc_fixed_estimates.append(mc_f.Q[state][action].item())
        
        td = QLearning(gamma=0.95, alpha=0.1, epsilon=0.1)
        td.train(GridWorld(), num_episodes=n_episodes)
        td_estimates.append(td.Q[state][action].item())
    
    mc_orig_estimates = np.array(mc_orig_estimates)
    mc_fixed_estimates = np.array(mc_fixed_estimates)
    td_estimates = np.array(td_estimates)
    
    fig, axes = plt.subplots(1, 2, figsize=(14, 5))
    
    axes[0].hist(mc_orig_estimates, bins=15, alpha=0.5,
                 label=f'MC orig (mean={mc_orig_estimates.mean():.2f}, var={mc_orig_estimates.var():.2f})', density=True, color='orange')
    axes[0].hist(mc_fixed_estimates, bins=15, alpha=0.5,
                 label=f'MC fixed (mean={mc_fixed_estimates.mean():.2f}, var={mc_fixed_estimates.var():.2f})', density=True, color='blue')
    axes[0].hist(td_estimates, bins=15, alpha=0.5,
                 label=f'TD (mean={td_estimates.mean():.2f}, var={td_estimates.var():.2f})', density=True, color='red')
    axes[0].axvline(true_q, color='black', linestyle='--', linewidth=2, label=f'True Q ≈ {true_q:.2f}')
    axes[0].set_xlabel('Q-value Estimate')
    axes[0].set_ylabel('Density')
    axes[0].set_title(f'Distribution of Q{target_state_action} Estimates')
    axes[0].legend(fontsize=8)
    axes[0].grid(True, alpha=0.3)
    
    axes[1].boxplot([mc_orig_estimates, mc_fixed_estimates, td_estimates],
                    labels=['MC (orig)', 'MC (fixed)', 'TD (Q-Learning)'])
    axes[1].axhline(true_q, color='black', linestyle='--', linewidth=2, label=f'True Q ≈ {true_q:.2f}')
    axes[1].set_ylabel('Q-value Estimate')
    axes[1].set_title('Variance Comparison')
    axes[1].legend()
    axes[1].grid(True, alpha=0.3, axis='y')
    
    plt.tight_layout()
    plt.show()
    
    print('\n' + '='*70)
    print('BIAS-VARIANCE ANALYSIS RESULTS')
    print('='*70)
    print(f'\nGround-truth Q-value: {true_q:.4f}')
    for label, ests in [('MC (last-visit, buggy)', mc_orig_estimates),
                         ('MC (first-visit, fixed)', mc_fixed_estimates),
                         ('TD (Q-Learning)', td_estimates)]:
        bias = ests.mean() - true_q
        print(f'\n{label}:')
        print(f'  Mean estimate: {ests.mean():.4f}')
        print(f'  Bias:          {bias:.4f}')
        print(f'  Variance:      {ests.var():.4f}')
    print('='*70)

compare_mc_td_variance()
```

## Summary & Key Takeaways

### Comparison Table

| Method | Bootstrapping | Updates | Bias | Variance | Policy Type | Best For |
|--------|---------------|---------|------|----------|-------------|----------|
| **Monte Carlo** | No | End of episode | Zero | High | Any | Episodic tasks |
| **SARSA** | Yes | Every step | Some | Low | On-policy | Safe exploration |
| **Q-Learning** | Yes | Every step | Some | Low | Off-policy | Finding optimal |

### When to Use Each Method?

**Monte Carlo:**
- Episodic problems only
- When unbiased estimates are critical
- Simple to implement

**SARSA (On-Policy):**
- Safety-critical applications
- Online learning scenarios
- When exploration risk matters
- Takes safer paths

**Q-Learning (Off-Policy):**
- Finding optimal policy
- Offline learning acceptable
- Can tolerate risky exploration
- Most popular in practice

### Key Insights:

1. **Bootstrapping Tradeoff**: TD methods trade bias for lower variance
2. **Convergence**: TD typically converges faster than MC
3. **On vs Off-Policy**: SARSA = safe, Q-Learning = optimal
4. **Q-functions for Control**: More practical than V-functions when model unknown
